```python
#step7:Exploring data using scatter plot for all possible four combinations
colors=['red','orange','blue']
species=['Iris-setosa','Iris-versicolor','Iris-virginica']
```

```python
for i in range(3):
    x =iris[iris['species'] == species[i]]
    plt.scatter(x['sepal_length'],x['sepal_width'],c=colors[i],label=species[i])
    plt.xlabel("sepal length")
    plt.ylabel("sepal width")
    plt.legend()
```

```python
#The scatter plot below is useful as there is very less overlapping
for i in range(3):
    x =iris[iris['species'] == species[i]]
    plt.scatter(x['petal_length'],x['petal_width'],c=colors[i],label=species[i])
    plt.xlabel("petal length")
    plt.ylabel("petal width")
    plt.legend()
```

```python
for i in range(3):
    x =iris[iris['species'] == species[i]]
    plt.scatter(x['sepal_length'],x['petal_length'],c=colors[i],label=species[i])
    plt.xlabel("sepal length")
    plt.ylabel("petal length")
    plt.legend()
```

```python
for i in range(3):
    x =iris[iris['species'] == species[i]]
    plt.scatter(x['sepal_width'],x['petal_width'],c=colors[i],label=species[i])
    plt.xlabel("sepal width")
    plt.ylabel("petal width")
    plt.legend()
```

```python
#step8:Replacing the string values in species columns to numerical value to observe correlation among data values
iris.replace({'species':{'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}},inplace=True)
iris.head()
```

```python
#step9:correlation in tabular format
iris.corr()
```

```python
#step10:Correlation in pictorial format
corr=iris.corr()
fig,ax=plt.subplots(figsize=(5,5))
sns.heatmap(corr, annot=True,ax=ax)
```

```python
#petal length and pental width have a high correlation of 0.96,so we can neglect one of them
#step11:Label Encoding for species column
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
iris['species']=l.fit_transform(iris['species'])
iris.head()
```

```python
#step12:model training
from sklearn.model_selection import train_test_split
x=iris.drop(columns=['species'])
y=iris['species']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
#applying logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#step13:Viewing the performance of the model
print("Accuracy:",model.score(x_test,y_test)
```
